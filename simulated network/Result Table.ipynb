```python
def style_specific_cell(x):
    global color_pos 
    color = 'background-color: red'
    df1 = pd.DataFrame('', index=x.index, columns=x.columns)
    for x in color_pos:
        df1.iloc[x[0],x[1]] = color
    return df1
```
